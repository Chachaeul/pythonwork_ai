        - 엔트로피 감소 = 불확실성 감소 = 순도 증가 = 정보 획득

        - 0~1의 수

### 복잡도 제어 : Prunning
( 가지치기 - 적당히 가지를 쳐야함 )
가지치기를 하는 이유?
            모든 terminal node의 순도가 100%인 상태를 Full tree라고 하는데요. 이렇게 Full tree를 생성한 뒤
            적절한 수준에서 terminal node를 결합해주어야 합니다. 왜냐하면 분기가 너무 많아서 학습데이터에
            과적합(overfitting)할 염려가 생기기 때문입니다
            
   - 사전 가지치기
   - 사후 가지치기
   - sklearn에서는 사후 가지치기만 지원
   - 매개변수
       max_depth 
       max_leaf_noads : 말단 노드의 최대 갯수
       min_samples_leaf : 리프노드가 되기 위한 최소한의 샘플 갯수
       min_samples_split : 노드가 분기할 수 있는 최소 샘플 갯수 (마지막 둘은 거의 비슷한 개념)
       max_features : 최적의 분할을위해 고려한 피처 갯수

```python
from sklearn.datasets import load_digits

mnist = load_digits()
mnist.data.shape
mnist.target

features, labels = mnist.data, mnist.target
```

```python
from sklearn.model_selection import cross_val_score

# 교차 검증 ex) cv=10 큰 덩어리를 10등분하여 1덩어리는 테스트용, 나머지는 train, 다른 1덩어리 테스트용, 나머지 train,,,...10번 하면 됨
def cross_validation(clf, features, labels) : 
    cv_scores = []
    
    for i in range(15) : 
        scores = cross_val_score(clf, features, labels, scoring="accuracy", cv=10 )
        
        cv_scores.append(scores.mean())
        
    return cv_scores
```

```python
df_cv_scores = cross_validation(DecisionTreeClassifier(), features, labels)
rf_cv_scores = cross_validation(DecisionTreeClassifier(), features, labels)
```

```python
cv_list = [["random_forest",rf_cv_scores],["dicision_tree",df_cv_scores]]
# df = pd.DataFrame({'random_forest' : rf_cv_scores, 'decision_tree' : df_cv_scores})
df = pd.DataFrame.from_dict(dict(cv_list))
df
df.plot()
```

## Gradient Boosting Regression Tree

    + 회귀와 분류에서 모두 사용
    + 이전 트리의 오차를 보완하는 방식으로 트리 생성(경사 하강법)
    + 강력한 사전 가지치기 사용
    + 랜덤 포레스트보다는 매개변수 설정에 민감하지만 잘 조정하면 더 높은 정확도 제공
    + learning_rate를 사용하여 오차 보정
    + XG Boost, LightGBM

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 42)
```

```python
# 깊이는 3. 트리 갯수는 100개, learning_rate=0.1
tree = GradientBoostingClassifier(random_state=0).fit(X_train, y_train)
print('훈련 정확도 : ', tree.score(X_train, y_train))
print('테스트 정확도 : ', tree.score(X_test, y_test))
```

```python
tree = GradientBoostingClassifier(random_state=0, max_depth=1).fit(X_train, y_train)
print('훈련 정확도 : ', tree.score(X_train, y_train))
print('테스트 정확도 : ', tree.score(X_test, y_test))
```

```python
tree = GradientBoostingClassifier(random_state=0, learning_rate=0.01).fit(X_train, y_train)
print('훈련 정확도 : ', tree.score(X_train, y_train))
print('테스트 정확도 : ', tree.score(X_test, y_test))
```
