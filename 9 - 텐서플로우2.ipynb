#### 경사하강 알고리즘을 이용한 가중치 합습

```python
# 데이터 준비
x_train = [1, 2, 3]
y_train = [1, 2, 3]

# 가설 준비 WX + b
W = tf.Variable(tf.random_normal([1]), dtype = tf.float32)
b = tf.Variable(tf.random_normal([1]), dtype = tf.float32)

# 어디서 시작할지 모르기 때문에 특정한 값대신 랜덤한 값을 입력
hypot = W * x_train

# 비용함수(손실함수)
cost = tf.reduce_mean(tf.square(hypot - y_train))

# 최저 비용학습을 위한 경사 하강 알고리즘(Gradient Desent)
alpha = 0.1
gradient = tf.reduce_mean((W * x_train - y_train) * x_train)
descent = W - alpha * gradient
update = W.assign(descent)  # W = decent


#-------------------------------- 텐서플로우의 그래프 작성 완료

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(21):
    _, c, w = sess.run([update, cost, W])
    print(step, c, w)
    
    
sess.close()
```

```python
filename_queue = tf.train.string_input_producer(
    ['data/data-01-test-score.csv'], shuffle=False, name='filename_queue')

reader = tf.TextLineReader()
key, value = reader.read(filename_queue)

# Default values, in case of empty columns. Also specifies the type of the
# decoded result.
record_defaults = [[0.], [0.], [0.], [0.]]
xy = tf.decode_csv(value, record_defaults=record_defaults)
xy
```

```python
train_x_batch, train_y_batch =  tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10)
train_x_batch
train_y_batch  # 요거 전부다 텐서다. 실행해야 10개씩 덩어리를 나눠서 순서대로 꺼내올수있다. 
```

```python
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis
hypothesis = tf.matmul(X, W) + b

# Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

#여기까지 그래프가 만들어진 결과다
```

```python
with tf.Session() as sess : 
    sess.run(tf.global_variables_initializer())
    
    
    coord = tf.train.Coordinator()  # 관리인을 지정해준다
    threads = tf.train.start_queue_runners(sess=sess, coord=coord) #학습직전에 준비해주시면 되는 코드
    
    
    for step in range(2001) : 
        x_batch, y_batch =  sess.run([train_x_batch, train_y_batch ]) #실행해야 10개씩 덩어리를 나눠서 순서대로 꺼내올수있다. 
        #10 개만 먼저 넘겨주는거다. 전체가 아니고. 원래는 2000번 반복. 이제는 하나의 에포크당 3번..?
        sess.run(train, fedd_dict={X : x_batch, y:y_batch})
        
    coord.request_stop()
    coord.join(threads)
    
    print("Your score will be ",
      sess.run(hypothesis, feed_dict={X: [[100, 70, 101]]}))

    print("Other scores will be ",
      sess.run(hypothesis, feed_dict={X: [[60, 70, 110], [90, 100, 80]]})) 
        
```

```python
# Logistic Regression

x_data = [[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]]
y_data = [[0], [0], [0], [1], [1], [1]]
```

```python
X = tf.placeholder(tf.float32, shape = [None, 2])
y = tf.placeholder(tf.float32, shape = [None, 1])

W = tf.Variable(tf.random_normal([2, 1]))
b = tf.Variable(tf.random_normal([1]))

hypot = tf.sigmoid(tf.matmul(X, W) + b)

cost = -tf.reduce_mean(y * tf.log(hypot) + (1 - y) * tf.log(1 - hypot))

train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(10000) :
    sess.run(train, feed_dict = {X : x_data, y : y_data})
```

```python

# 정확도 측정

pred = tf.cast(hypot > 0.5, dtype = tf.float32)
sess.run(pred, feed_dict = {X : x_data, y : y_data})
accuracy = tf.reduce_mean(tf.cast(tf.equal(pred, y), dtype = tf.float32))

h, p, a = sess.run([hypot, pred, accuracy],
                   feed_dict = {X : x_data, y : y_data})

print('가설 :', h, '\n예측 :', p, '\n정확도 :', a)
```

## 사례1 : 동물 분류

```python
xy = np.loadtxt("data/data-04-zoo.csv", delimiter=",",dtype=np.float32)
xy.shape

x_data = xy[:,0:-1]  # 모든 행, 맨마지막 열 빼고
y_data = xy[:,[-1]] # 모든행, 맨마지막 열만
```

```python
X = tf.placeholder(tf.float32, shape=[None, 16])
y = tf.placeholder(tf.int32, shape=[None, 1])

y_one_hot = tf.one_hot(y ,7)
y_one_hot = tf.reshape(y_one_hot, [-1, 7])

W = tf.Variable(tf.random_normal([16, 7]))
b = tf.Variable(tf.random_normal([7]))

logit = tf.matmul(X, W) + b
hypot = tf.nn.softmax(logit)
cost = tf.nn.softmax_cross_entropy_with_logits(logits=logit, labels=y_one_hot)

train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

pred = tf.argmax(hypot,1)
correct = tf.equal(pred, tf.argmax(y_one_hot,1))
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

for step in range(200):
    sess.run(train, feed_dict={X:x_data, y:y_data})
    
    if step%10 ==0  :
        loss, acc= sess.run([tf.reduce_mean(cost), accuracy], feed_dict={X:x_data, y:y_data})
        
        print("step:{:5}\tLoss:{:.3f}\tAcc:{}".format(step,loss,acc))
```

### 사례2 : MNIST 숫자 이미지

```python
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets("data/MNIST_data/", one_hot=True)
```

```python
type(mnist)
dir(mnist)
mnist.train.labels   #one_hot=True 옵션때문에 데이터가 원핫인코딩 형식으로 갖춰져있다. 원래는(0-9)
mnist.train.num_examples
```

```python
X = tf.placeholder(tf.float32, shape=[None, 28*28])
y = tf.placeholder(tf.float32, shape=[None, 10])

W = tf.Variable(tf.random_normal([784, 10]))
b = tf.Variable(tf.random_normal([10]))

logit = tf.matmul(X, W) + b
hypot = tf.nn.softmax(logit)
cost = tf.nn.softmax_cross_entropy_with_logits(logits=logit, labels=y)

train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

correct = tf.equal(tf.arg_max(hypot, 1), tf.argmax(y, 1))
accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

sess = tf.Session()
sess.run(tf.global_variables_initializer())

batch_size = 100

for epoch in range(15):  #반복 횟수는 15이지만 실제 횟수는 어마어마 하겠죠? #55000rofmf 100씩 자르겠다. 총 몇번의 배치를 할 것인지 생각.
      total_batch= int(mnist.train.num_examples / batch_size)
    
      for i in range(total_batch):
        batch_xs, batch_ys = mnist.train.next_batch(batch_size)
        # 100개, 그다음 100개, .. 순차적으로 가지고 올수 있게 해줌
        # total 배치만큼 위 활동을 다 해주면 epoch 1번. 따라서 실제로는 많은 반복을 도는 것
        

        _, loss, acc = sess.run([train, tf.reduce_mean(cost), accuracy], feed_dict={X:batch_xs, y:batch_ys})
        #train, 비용, 정확도가 어떻게 학습되고 있는지 확인
        #placeholder에 한번에 550000개가 아니라 100개씩 넘겨주므로 메모리 부담 없음.
        
        print("step:{:5}\tLoss:{:.3f}\tAcc:{}".format(epoch,loss,acc))
        
        
```
